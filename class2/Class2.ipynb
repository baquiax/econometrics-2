```python
from time import time
b0 = 5
b1 = 2

x = np.arange(1, 10000000)

y = np.empty_like(x)
y
```

```python
start = time()
for index, value in enumerate(x):
    y[index] = b0 + b1 * value
end = time()

print(f"Elapsed time: {end - start}")
y
```

```python
start = time()
y = b0 + b1 * x
end = time()

print(f"Elapsed time: {end - start}")
y
```

### Solución con NumPy

```python
vect1 = np.ones_like(x).reshape(-1, 1)
x = np.reshape(x, (-1, 1))
A = np.hstack([x, vect1])
betas = np.array([b1, b0])

start = time()
y1 = np.matmul(A, betas)
end = time()
print(f"Elapsed time: {end - start}")
y1
```

```python
x = np.reshape(x, (-1, 1))
vect1 = np.ones_like(x)
A = np.hstack([x, vect1])
betas = np.array([b1, b0])

start = time()
y1 = np.matmul(A, betas)
end = time()
print(f"Elapsed time: {end - start}")
y1
```

#### Filtros

```python
df3 = pd.DataFrame(
    {
        'A': np.random.randint(1, 21, 10),
        'B': np.random.randint(1, 21, 10),
        'Et': np.array([True, True, True, False, False, False, True, False, True, False])
    },
    index=np.arange(1, 11, 1)
)
df3
```

```python
df3[df3["A"] > 10]
```

```python
df3[(df3["A"] > 10 ) & (df3["B"] < 5)]
```

```python
df3[(df3["A"] < 5 ) | (df3["B"] > 10)]
```

```python
df3[(df3["Et"] == False )]
```

```python
df3[(df3["Et"] == False ) & (df3["A"] > 5)].loc[:, ["A"]]
```
