## E:1

Haga un programa en Python que reciba de parámetro un número de segundos y que despliegue en pantalla, las horas, minutos y segundos que representan. Por ejemplo:

1. Si el número de segundos recibidos fuera: 2743 entonces su programa debería desplegar: Horas: 0, Minutos: 45 , Segundos: 43
2. Si el número de segundos recibidos fuera 5871: Horas: 1, Minutos: 37, Segundos: 51

````python
seconds_in_hour=60*60
seconds_in_minute=60

def print_description(seconds = 0):
  if seconds < 0:
    print(f"The number of seconds given {seconds} MUST be non-negative \n")
    return

  hours = seconds / seconds_in_hour
  remaining_seconds = seconds % seconds_in_hour

  minutes = remaining_seconds / seconds_in_minute
  remaining_seconds = remaining_seconds % seconds_in_minute

  print(f"{seconds} second(s) could be represented as: {int(hours)} hour(s), {int(minutes)} minute(s) and {int(remaining_seconds)} second(s)")

 
print_description(2743)
print_description(5871)
print_description(0)
print_description(-100)
````

## E:2

Haga una función en Python que dado un número entre `1` y `20`, y un carácter cualquiera, despliegue cuatro patrones triangulares.

Por ejemplo si el usuario ingresara el número `4` y el súmbolo `*` el programa deberá desplegar lo siguiente:

```
*       ****      ****      *     
**       ***      ***      **
***       **      **      *** 
****       *      *      ****
```

````python
space_size=8

def print_triangles(width = 0, char = "*"):
  separator=' ' * space_size

  if width < 1 or width > 20:
    print(f"width should be a number between 1 and 20, but it was given {width}")
    return

  for i in range(1, width + 1):
    leading=char*i
    trailing=leading

    middle = char*(width + 1 - i) 

    print(f"{leading}{separator}{middle}{separator}{middle}{separator}{trailing}")

print_triangles(0)
print_triangles(4)
print_triangles(4, "-")
print_triangles(-1)
print_triangles(20, "-")
````

## E:3

Haga un programa en Python que la fecha de nacimiento de una persona (puede ingresar por separado cada valor, es decir pedir primero día, luego mes y luego año), despliegue la edad de la persona en años, meses y días. Su resultado puede ser aproximado.

````python
import datetime;

def get_int_input():
    value_as_string = input()
    try:
      return int(value_as_string)
    except:
      return 0

def try_get_positive_input(message, max):
  while True:
    print(message)
    value = get_int_input()

    if value > 0 and value <= max:
      return value

def calculate_time_life():
  day = try_get_positive_input("Please, insert your birth day (e.g 31)", 31)
  month = try_get_positive_input("Please, insert your birth month (e.g 12)", 12)
  year = try_get_positive_input("Please, insert your birth year (e.g 1993)", 2023)

  parsed_date = datetime.datetime(year, month, day)

  diff = datetime.datetime.now() - parsed_date

  elapsed_years = diff.days / 365
  elapsed_days = diff.days % 365
  elapsed_hours = elapsed_days % 24
  

  print(f"Your a1ge is: {int(elapsed_years)} years, {elapsed_days} days, {elapsed_hours} hours")
calculate_time_life()
````

Escriba una función en Python que dada una tupla de elementos (números o letras) esta se convierta a un string. Tomar nota del siguiente ejemplo:

1. Si la entrada fuera: (10, 20, 40, 5, 70)
  
  1.1 La salida debería ser: 102040570

````python
def tuple_to_string(tuple = ()):
  result = ""
  for e in tuple:
    result += str(e)
  return result

print(tuple_to_string((10, 20, 40, 5, 70)))
print(tuple_to_string(()))
````

## E:5

Cree un programa en Python que dada una lista de tuplas, elimine las tuplas que están vacías. Tomar nota del siguiente ejemplo:

1. Si la lista fuera: `[(), (), ('X'), ('a', 'b'), ('a', 'b', 'c'), ('d')]`

  1.1 `El resultado deber´ıa ser: [('X',), ('a', 'b'), ('a', 'b', 'c'), 'd']`


````python
def remove_empty(list=[]):
  result = []

  for e in list:
    if len(e) == 0:
      continue
    
    result.append(e)

  return result

print(remove_empty([(), (), ('X',), ('a', 'b'), ('a', 'b', 'c'), ('d')]))
````

## E:6

Cree una función en Python que dada una tupla de tuplas con números, produzca una tupla con el promedio de cada tupla individual. Tomar nota del siguiente ejemplo:

1. Si la tupla de entrada fuera: `((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32))`

  1.1 El resultado debería ser: `(10.5, 44, 58)`

````python
def tuples_avg(tuple_list = ()):
  result = []

  for t in tuple_list:
    sum = 0
    for number in t:
      sum += number

    result.append(sum/ len(t))

  return tuple(result)


print(tuples_avg(((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32))))
````

## E:7

Escribir un programa que gestione las facturas pendientes de cobro de una empresa. Las facturas se almacenarán en un diccionario donde la llave de cada factura será el número de factura (debe utilizar un numero de 4 dígitos) y el valor el coste de la factura. El programa debe preguntar al usuario si quiere añadir una nueva factura, pagar una existente o terminar.

Si desea añadir una nueva factura se preguntará por el número de factura y su coste y se añadirá al diccionario, para confirmar que la factura se ingreso correctamente el programa debe desplegar el diccionario completo. 

Si se desea pagar una factura se preguntará por el número de factura y se eliminará del diccionario. 

Después de cada operación el programa debe mostrar en pantalla la cantidad cobrada hasta el momento y la cantidad pendiente de cobro. Puede utilizar otras estructuras de datos como apoyo si lo considera necesario

````python
import math

def get_string_input(message, min_length = 0, max_length = math.inf):
  while True:
    print(message)

    value_as_string = input()
    if len(value_as_string) >= min_length and len(value_as_string) <= max_length:
      return value_as_string



def get_float_input(message, min_value = 0.0, max_value = math.inf):
  while True:
    print(message)

    value_as_string = input()

    try:
      value = float(value_as_string)

      if value >= min_value and value <= max_value:
        return value

    except:
      continue

def get_int_input(message, min_value = 0, max_value = math.inf):
  while True:
    print(message)

    value_as_string = input()

    try:
      value = int(value_as_string)

      if value >= min_value and value <= max_value:
        return value

    except:
      continue

def print_invoices(invoices = {}):
  print("\n\nHere the current list of invoices:\n")
  for x in invoices:
    print(f"{x}\t{invoices[x]}")
  
  print("")

def main():
  invoices = {}

  menu = '''
What do you want to do?

1. Add invoice
2. Pay invoice
0. Exit'''

  while True:
    input = get_int_input(menu, min_value=0, max_value=2)

    if input == 0:
      print_invoices(invoices)
      return

    elif input == 1:
      print("Sure! Please provide the id and the amount for yout new invoice")

      id = get_string_input("Please, insert the invoice id (4 chars): ", min_length=4, max_length=4)

      if id in invoices:
          print("Duh! That id already exists. Sending you to the menu.")
          continue

      amount = get_float_input("Now, please insert the amount (e.g 1.0)", min_value=1)

      invoices[id] = amount

      print(f"Done! I just registered your invoice {id}")

      print_invoices(invoices)

    elif input == 2:
      print("Of course! I'll need some info to get the invoice's payment")

      id = get_string_input("Could you gimme the invoice id (4 chars): ", min_length=4, max_length=4)

      if id not in invoices:
          print("Sorry! That id is not present in my database. Sending you to the menu.")
          continue

      invoices.pop(id)

      print(f"Done! I just removed the invice id {id}")

      print_invoices(invoices)
    

main()
````

Haga una función en Python, que cree un set con todas las cartas de una baraja de Pocker utilizando el siguiente formato: `Tipo-Numero`.

Utilice las letras `T` para tréboles, `C` para
corazones, `P` para picas y `E` para espadas. Para los números utilice la numeración del `1` al
`13`. Por ejemplo `E13` - sería Rey de Espadas, `P01` sería As de Picas, `T05` sería cinco de Tréboles.

Posteriormente cree una función que dado el conjunto anterior cree dos sets nuevos de cartas, cada conjunto debe tener una cantidad variable de elementos, puede utilizar números aleatorios enteros para definir la longitud de los conjuntos.

Adicionalmente, debería calcular la probabilidad de la intersección de ambos conjuntos y demuestre si son eventos independientes con la regla del producto.

Para concluir, calcule probabilidad de la unión de ambos conjuntos y demuestre que se cumple
la regla de la adición.

````python
import random

def build_cards():
  cards = set()

  for i in range(1, 14):
    cards.add(f"T{i:02d}")
    cards.add(f"E{i:02d}")
    cards.add(f"C{i:02d}")
    cards.add(f"P{i:02d}")
  
  return cards


def get_random_subset(cards = set()):
  cards_as_list = list(cards)

  subset = set()
  subset_length = random.randint(1, len(cards)) # at least 1

  for i in range(0, subset_length):
    index_to_remove = random.randint(0, len(cards_as_list) - 1)
    card = cards_as_list.pop(index_to_remove)

    subset.add(card)

  
  return subset



def main():
  cards = build_cards()

  A = get_random_subset(cards)
  B = get_random_subset(cards)


  A_probability = round(len(A)/52, 2)
  print(f"A = {A}")
  print(f"P(A) = {A_probability} ({len(A)})\n")

  B_probability = round(len(B)/52, 2)
  print(f"B = {B}")
  print(f"P(B) = {B_probability} ({len(B)})\n")

  # Product rule
  print(f"\n## is A independant of B? We need to proof that P(A|B) = P(A)\n")

  intersection = A.intersection(B)
  intersection_probability = round(len(intersection)/52, 2)
  A_given_B_probability = round(intersection_probability / B_probability, 2)

  print(f"A n B = {intersection} ({len(intersection)})")
  print(f"P(A n B) = {intersection_probability}\n")

  print(f"P(A|B) = P(A n B) / P(B)")
  print(f"P(A|B) = {intersection_probability} / {B_probability} = {A_given_B_probability}\n")
  print(f"P(A|B)={A_given_B_probability} is equal to P(A)={A_probability}? = {A_probability == A_given_B_probability}")


  # Sum rule
  print(f"\n## Let's calculate P(A u B)")

  union = A.union(B)
  union_probability = round(len(union)/52, 2)
  print(f"P(A u B) = {union_probability}")

  print(f"\nif A is mutually exclusive of B then P(A u B) must be P(A) + P(B), let's test it out")
  A_plus_B_probability = A_probability + B_probability

  print(f"is P(A)={A_probability} + P(B)={B_probability} equal to P(A u B)={union_probability}? {A_plus_B_probability == union_probability}")
main()
````
